```python
import pandas as pd

#1 reading the csv file
df = pd.read_csv('acc_players-2324F.csv', header=1)
df.head()
```

```python
# 2 basic analysis
total_points = df['PTS'].sum()
print(f'Total points scored by all players combined: {total_points}', 'points', '\n')

most_minutes_player = df.loc[df['MP'].idxmax()]
print(f'Player with the most minutes: {most_minutes_player["Player"]}', '\n')

top_5_rebounders = df.nlargest(5, 'TRB')
print('Top 5 players in terms of total rebounds:')
print(top_5_rebounders[['Player', 'TRB']])
```

```python
# 3 player filtering
filtered_df = df[df['MP'] > 500]

highest_assists_player = filtered_df.loc[filtered_df['AST'].idxmax()]
print(f'Player with the highest total assists: {highest_assists_player["Player"]}', '\n')

top_3_assist_leads = filtered_df.nlargest(3, 'AST')
print('Top 3 Assist Leads in the League:')
print(top_3_assist_leads[['Player', 'AST']], '\n')

top_3_shot_blockers = filtered_df.nlargest(3, 'BLK')
print('Top 3 Shot Blockers:')
print(top_3_shot_blockers[['Player', 'BLK']], '\n')
```

```python
# 4 school-based analysis
school_points = df.groupby('School')['PTS'].sum().reset_index()
print('Total points scored by each school:')
print(school_points, '\n')

school_assists = df.groupby('School')['AST'].sum().reset_index()
print('Total assists by each school:')
print(school_assists, '\n')

top_3_schools = school_points.nlargest(3, 'PTS')
print('Top 3 schools by total points scored:')
print(top_3_schools)
```

```python
# bonus 
import matplotlib.pyplot as plt

top_5_scorers = df.nlargest(5, 'PTS')
plt.bar(top_5_scorers['Player'], top_5_scorers['PTS'])
plt.xlabel('Player')
plt.ylabel('Total Points Scored')
plt.title('Top 5 Players by Total Points Scored')
plt.show()


plt.scatter(df['FG%'], df['WS'])
plt.xlabel('Field Goal Percentage (FG%)')
plt.ylabel('Win Shares (WS)')
plt.title('Correlation between FG% and WS')
plt.show()

correlation_coefficient = df['FG%'].corr(df['WS'])
print(f'Correlation coefficient between FG% and WS: {correlation_coefficient}')
```
