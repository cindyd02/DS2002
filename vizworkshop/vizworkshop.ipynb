```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

airbnb_data = pd.read_csv('airbnb_dataset.csv')
iris_data = pd.read_csv('iris_dataset.csv')

plt.figure(figsize=(8, 6))
plt.scatter(airbnb_data['Longitude'], airbnb_data['Latitude'], c=airbnb_data['Price'], cmap='viridis', s=50, alpha=0.7)
plt.colorbar(label='Price')
plt.title('Airbnb Listings by Location')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()

plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_data, x='sepal length (cm)', y='sepal width (cm)', hue='target', palette='viridis', s=100)
plt.title('Sepal Dimensions by Target')
plt.show()

fig = px.scatter(
    airbnb_data,
    x='Longitude',
    y='Latitude',
    color='Price',
    size='Rating',
    hover_data=['Price', 'Rating'],
    title='Airbnb Locations with Price and Rating'
)
fig.show()

iris_data_renamed = iris_data.rename(columns={
    'sepal length (cm)': 'Sepal Length',
    'sepal width (cm)': 'Sepal Width',
    'petal length (cm)': 'Petal Length',
    'petal width (cm)': 'Petal Width'
})

fig = px.scatter_matrix(
    iris_data_renamed,
    dimensions=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'],
    color='target',
    title='Iris Dataset Scatter Matrix'
)

fig.update_layout(
    title=dict(x=0.5),  
    font=dict(size=10), 
    autosize=False,
    width=1000, 
    height=1000 
)

fig.show()

```
